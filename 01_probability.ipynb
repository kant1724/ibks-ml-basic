```
import scipy as sp
import scipy.stats
from scipy.special import beta, gamma
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Matplotlib -> 데이터를 시각화 해주는 툴(차트, 플롯)

### Probability Density Function

```
# 평균, 표준편차 (정규분포)
rv = sp.stats.norm(loc=0, scale=2)
# -8 ~ 8 까지의 범위를 100조각으로 균등분할 한 값
xx = np.linspace(-8, 8, 100)
pdf = rv.pdf(xx)
plt.plot(xx, pdf)
plt.title('pdf')
plt.xlabel('x')
plt.ylabel('p(x)')
plt.show()
```

###Cumulative Distribution Function

```
rv = sp.stats.norm(loc=0, scale=1)
xx = np.linspace(-8, 8, 100)
# 누적분포함수
cdf = rv.cdf(xx)
plt.plot(xx, cdf)
plt.title('cdf')
plt.xlabel('x')
plt.ylabel('F(x)')
plt.show()
```

###Binomial Distribution

```
# 전체시행횟수
N = 10
# 이산 확률
mu = 0.5
rv = sp.stats.binom(N, mu)
# 숫자를 0부터 N 까지 순서대로 배열한다.
xx = np.arange(N + 1)
print(xx)
# pmf(probability mass function) - 확률질량함수
plt.bar(xx, rv.pmf(xx), align='center')
plt.xlabel('x')
plt.ylabel('P(x)')
plt.title('pmf')
plt.show()
```

```
# 이산확률변수를 가지는 행위를 N번 독립시행했을 때 어떠한 결과가 나오는지 그래프로 표현
x = rv.rvs(1000)
sns.countplot(x)
plt.title('count')
plt.xlabel('x')
plt.show()
```

###Beta Distribution

```
x = np.linspace(.01, .99, 99)
# 베타분포의 파라미터
a = 0.5
b = 0.5
# 베타분포
y = (1 / beta(a, b)) * x ** (a - 1) * (1 - x) ** (b - 1)

plt.figure(figsize=(12, 8))
plt.plot(x, y)
plt.xlabel('X')
plt.ylabel('P(X)')
plt.title('Beta Distribution')
plt.grid()
plt.show()
```
