```python
import scipy as sp
import scipy.stats
from scipy.special import beta, gamma
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Probability Density Function

```python
# 평균 0, 표준편차 4
rv = sp.stats.norm(loc=0, scale=4)
xx = np.linspace(-8, 8, 100)
pdf = rv.pdf(xx)
plt.plot(xx, pdf)
plt.title('pdf')
plt.xlabel('x')
plt.ylabel('p(x)')
plt.show()
```

### Cumulative Distribution Function

```python
rv = sp.stats.norm(loc=0, scale=4)
xx = np.linspace(-8, 8, 100)
cdf = rv.cdf(xx)
plt.plot(xx, cdf)
plt.title('cdf')
plt.xlabel('x')
plt.ylabel('F(x)')
plt.show()
```

### Binomial Distribution

```python
N = 10
mu = 0.5
rv = sp.stats.binom(N, mu)
xx = np.arange(N + 1)
plt.bar(xx, rv.pmf(xx), align='center')
plt.xlabel('x')
plt.ylabel('P(x)')
plt.title('pmf')
plt.show()
```

```python
x = rv.rvs(100)
sns.countplot(x)
plt.title('count')
plt.xlabel('x')
plt.show()
```

### Beta Distribution

```python
x = np.linspace(.01, .99, 99)
a = 0.5
b = 0.5
y = (1 / beta(a, b)) * x ** (a - 1) * (1 - x) ** (b - 1)

plt.figure(figsize=(12, 8))
plt.plot(x, y)
plt.xlabel('X')
plt.ylabel('P(X)')
plt.title('Beta Distribution')
plt.grid()
plt.show()
```
