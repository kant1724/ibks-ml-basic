```python
# 평균 0, 표준편차 4
rv = sp.stats.norm(loc=0, scale=4)
xx = np.linspace(-8, 8, 100)
pdf = rv.pdf(xx)
plt.plot(xx, pdf)
plt.title("pdf ")
plt.xlabel("$x$")
plt.ylabel("$p(x)$")
plt.show()
```
